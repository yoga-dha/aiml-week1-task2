# Student Performance Data Visualization

The objective of this task is to create visualizations from the student performance dataset and identify trends, patterns, and relationships affecting GPA.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Student_performance_data _.csv")
df.head()
```

The dataset was loaded successfully for visualization.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['GPA'], bins=10, kde=True)

plt.title("Distribution of GPA")
plt.xlabel("GPA")
plt.ylabel("Frequency")

plt.show()
```

The graph displays the distribution of GPA values among students. Most students are concentrated around the average GPA range.

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Gender', data=df)

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

The graph shows the number of students in each gender category. The dataset appears reasonably balanced.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=df)

plt.title("Study Time vs GPA")
plt.xlabel("Study Time Weekly")
plt.ylabel("GPA")

plt.show()
```

Students who spend more time studying generally tend to achieve higher GPA values. A positive relationship can be observed.

```python
corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='Blues')

plt.title("Correlation Heatmap")

plt.show()
```

The heatmap displays correlations among numerical variables. Positive and negative relationships can be identified easily.

Strongest Positive Correlation:
StudyTimeWeekly and GPA

Weak Correlation:
Age and GPA

Negative Correlation:
Absences and GPA

Note:
Correlation does not imply causation.

1. Students with higher study time generally achieve better GPA.

2. Absences may negatively affect academic performance.

3. GPA distribution is balanced across the dataset.

4. Academic performance is influenced by multiple factors.

5. Study habits play an important role in student success.

The visualizations helped identify important patterns in student performance. Study habits, attendance, and support-related variables appear to influence GPA. Correlation analysis confirmed relationships among academic and behavioral factors.
